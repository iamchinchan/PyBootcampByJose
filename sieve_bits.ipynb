```python
MAX = 10**9
# 1. Initialize bitmask (All 1s)
isPrime = (1 << (MAX + 1)) - 1

def sieve():
    global isPrime
    # 2. Kill 0 and 1
    isPrime &= ~0b11 
    
    # 3. Process 2 separately to enable 2*factor jump
    for multiple in range(4, MAX + 1, 2):
        isPrime &= ~(1 << multiple)
    
    # 4. Process odd factors
    for factor in range(3, int(MAX**0.5) + 1, 2):
        # 5. Right-shift check is faster than creating a mask
        if (isPrime >> factor) & 1:
            # 6. Jump by 2*factor to skip even multiples (already 0)
            for multiple in range(factor * factor, MAX + 1, 2 * factor):
                isPrime &= ~(1 << multiple)

sieve()
```

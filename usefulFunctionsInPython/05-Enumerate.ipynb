# enumerate()

In this lecture we will learn about an extremely useful built-in function: enumerate(). Enumerate allows you to keep a count as you iterate through an object. It does this by returning a tuple in the form (count,element). The function itself is equivalent to:

    def enumerate(sequence, start=0):
        n = start
        for elem in sequence:
            yield n, elem
            n += 1

## Example

```python
lst = ['a','b','c']

for number,item in enumerate(lst):
    print(number)
    print(item)
```

```python
en = enumerate(lst)
print(en)
print(list(en))
print(list(en)) #exhausts
```

```python
fruits = ['apple', 'banana', 'cherry']

# Default starting index is 0
for index, fruit in enumerate(fruits):
    print(index, fruit)

# Output:
# 0 apple
# 1 banana
# 2 cherry
```

```python
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit) # Starts at 1
```

It's a subtle but important distinction:
enumerate is the function (specifically a class constructor).
The thing it returns is the iterator.
Think of it like a factory:
The Function (enumerate): This is the machine. You feed a list into it.
The Object: This is the product that comes out of the machine. That product is the iterator.
The Proof:


```python
fruits = ['apple', 'banana']

# 1. Calling the function
result = enumerate(fruits) 
print(result)
# enumerate object
print(type(result)) 
# Output: <class 'enumerate'> 

# 2. Testing if the result is an iterator
print(iter(result) is result) 
# Output: True (It returns self! It's an iterator!)

# 3. Using it
print(next(result)) # (0, 'apple')
print(list(result))
print(list(result)) #exhausts
```

enumerate() becomes particularly useful when you have a case where you need to have some sort of tracker. For example:

```python
print(lst)
for count,item in enumerate(lst):
    if count >= 2:
        break
    else:
        print(item)
```

enumerate() takes an optional "start" argument to override the default value of zero:

```python
months = ['March','April','May','June']

list(enumerate(months,start=3))
```

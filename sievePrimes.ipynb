```python
test_inputs = [
    -5, 0, 1, 2, 3, 4, 9, 11,           # Edge Cases
    100, 1000, 7919, 10000, 100000,      # Small to Medium Scale
    1000000, 10000000, 100000000,        # Large Scale (10^6, 10^7, 10^8)
    1000000000                          # Boss Level (10^9)
]

# Matching Expected Answers (pi(x) values)
test_answers = [
    0, 0, 0, 1, 2, 2, 4, 5,              # Edge Answers
    25, 168, 1000, 1229, 9592,           # Scale Answers
    78498, 664579, 5761455,              # Large Scale Answers
    50847534                             # Boss Level Answer
]
```

```python
MAX = 10**9
isPrime = [True]*(MAX+1)
def sieve():
  isPrime[0]=isPrime[1]=False
  isPrime[4::2]=[False]*((MAX-4)//2+1)
  for i in range(3,int(MAX**0.5)+1,2):
    if isPrime[i]:
      start=i*i
      step=2*i
      isPrime[start:MAX+1:step]=[False]*((MAX-start)//step+1)
sieve()
```

```python
def countPrime(num):
  if num<0: return 0
  return sum(isPrime[0:num+1])
```

```python
answers = [countPrime(num) for num in test_inputs]
```

```python
answers==test_answers
```

```python
with open("Working_Business_Proposal.pdf","rb") as read_file_object:
  pdf_reader = pypdf.PdfReader(read_file_object)
  page_two = pdf_reader.pages[2]
  #3rd page is at index 2s
  with open("3rd_page_file.pdf",mode="wb") as write_file_object:
    pdf_writer =  pypdf.PdfWriter()
    print(f"pdf writer is  :{pdf_writer}")
    pdf_writer.add_page(page_two)
    pdf_writer.write(write_file_object)
```

```python
with open("pdf_first_page.pdf",mode="wb") as write_file_object:
    pdf_writer =  pypdf.PdfWriter()
    print(f"pdf writer is  :{pdf_writer}")
    print(f"Adding page reference")
    print(page_one)
    print(type(page_one))
    pdf_writer.add_page(page_one)
    print("page reference added to writer object")
    pdf_writer.write(write_file_object)
```

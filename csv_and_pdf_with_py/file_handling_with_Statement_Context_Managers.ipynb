# With Statement Context Managers

When i open a file using `f = open('test.txt')`, the file stays open until i specifically call `f.close()`.  Should an exception be raised while working with the file, it remains open. This can lead to vulnerabilities in my code, and inefficient use of resources.

A context manager handles the opening and closing of resources, and provides a built-in `try/finally` block should any exceptions occur.

### Standard `open()` procedure, with a raised exception:


```python
p = open('oops.txt','a')
p.readlines()
p.close()
```

Let's see if we can modify our file:

```python
p.write('add more text')
```

Ouch! I may not have wanted to do that until I traced the exception! Unfortunately, the exception prevented the last line, `p.close()` from running. Let's close the file manually:

```python
p.close()
```

### Protect the file with `try/except/finally`

A common workaround is to insert a `try/except/finally` clause to close the file whenever an exception is raised:


```python
p = open('oops.txt','a')
try:
    p.readlines()
except:
    print('An exception was raised!')
finally:
    p.close()
```

Let's see if i can modify our file this time:

```python
p.write('add more text')
```

Excellent! Our file is safe.

### Save steps with `with`

Now i'll employ my context manager. The syntax follows `with [resource] as [target]: do something`

```python
with open('oops.txt','a') as p:
    p.readlines()
```

Can i modify the file?

```python
p.write('add more text')
```

Great! With just one line of code i've handled opening the file, enclosing my code in a `try/finally` block, and closing my file all at the same time.

```python
import smtplib
```

```python
smtp_object = smtplib.SMTP("smtp.gmail.com",587)
#587 for tls connection
#created the smtp object
```

```python
smtp_object
```

```python
# not establish the connection and confirm it by output
smtp_object.ehlo()
```

```python
#250 output so ok
```

```python
#lets start tls
smtp_object.starttls()
```

```python
#220 to ready lets now login
```

```python
import getpass
```

```python
email = getpass.getpass("Enter Email: ")
password = getpass.getpass("Enter Password: ")
```

```python
smtp_object.login(email,password)
```

```python
# as success , now lets create the attributes for sending message
```

```python
from_email_adress = email
to_email_adress = "goyaljatin344@gmail.com"
subject = "Learning to send emails with Python scripts"
message = "This is a mail to myself from my other Gmail id: and I am trying to learn python scripts for sending and receiving mails throgh this and on gmail using gmail domain name and smtp protocol"
msg = f"Subject : {subject}\n\n{message}"
```

```python
smtp_object.sendmail(from_email_adress,to_email_adress,msg)
```

```python
#{} means success
```

```python
smtp_object.quit()
```

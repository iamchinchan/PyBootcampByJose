```python
def display_rttt(ttt):
  print("This is  a reference table to select a position from: ")
  for r in range(len(ttt)):
    print(" | ".join(f"{r*len(ttt)+c+1:2}" for c in range(len(ttt))))
    if r!=len(ttt)-1:
      print("___"*(len(ttt)+2))
```

```python
def display_ttt(ttt):
  print("Your tic tac tow Now looks like this: ")
  print(f"\n{'__'*(len(ttt)+2)}\n".join(" | ".join(f"{c}" for c in r) for r in ttt))

  
```

```python
def assign_player_codes():
  choice="wrong"
  while choice.upper() not in ["X","O"]:
    # when choice is either x/X/o/O it will break out of loop
    choice = input("Player 1 choose which you wanna pick as your symbol: X or O")
    if choice.upper() not in ["X","O"]:
      print("Please try again")
    else:
      pass
  print(f"Player 1 has choosen \"{choice.upper()}\"")
  return {1:"X",2:"O"} if choice.upper()=="X" else {1:"O",2:"X"}
  
```

```python
def get_unmarked_position(ttt,sides):
  choice=False
  while not choice:
    position =input(f"Please choose a spot to mark between 1- {sides*sides} which is not marked already")
    if position.isdigit():
      position = int(position)
      if position>=1 and position<=sides*sides:
        r = (position-1)//sides
        c = (position-1)%sides
        if(ttt[r][c]==" "):
          return(r,c)
        else:
          print(f"{position}: position already occupied, please choose unoccupied position to play further")
      else:
        print(f"please type a digit between {1} to {sides*sides}")
    else:
      print(f"please enter a digit")
```

```python
def won_or_not(ttt,r,c,playerCode):
  #check for row wise:
  for cl in range(len(ttt)):
    if ttt[r][cl]!=playerCode:
      break
  else:
    #not breaked out of upper loop
    return True
  
  #check for columnwise
  for rw in range(len(ttt)):
    if ttt[rw][c]!=playerCode:
      break
  else:
    #if not break then all same as playerCode now
    return True
  
  if r==c:
    #check left diagonal
    for i in range(len(ttt)):
      if ttt[i][i]!=playerCode:
        break
    else:
      #if not break then it means all same as playercoder and player won
      return True
    
  if r+c==len(ttt)-1:
    #means its a right diagonal
    for rw in range(len(ttt)):
      cl = len(ttt)-1-rw
      if ttt[rw][cl]!=playerCode:
        break
    else:
      return True
      #right diagonal is completely similart to playucode
    
  return False
# must as we just breaked through if we did not find a win condiiton, so we have to return false
```

```python
def switch_players(playerNo):
  return 1 if playerNo==2 else 2
```

```python
def get_sides():
  while True:
    sides = input("Enter number of sides of tic tac toe: ")
    if sides.isdigit():
      sides=int(sides)
      if 3<= sides <=6:
        return sides
      else:
        print(f"Please enter sides > 2 and sides <7 to play properly")
        continue
    else:
      print("please eneter a digit: ")
      continue
```

```python
def play_ttt():
  # clear_output()
  
  print("Welcome to Tic Tac Toe! Game: This Game is played by two players:")
  sides=get_sides()
  print(f"User has selected to play tic tac toe in {sides} sides")
  #its a side*side tic tac toe
  positions_left=sides*sides
  #total positions left to place marker upon, starting with all of them avaialble

  ttt  = [[" " for _ in range(sides)] for _ in range(sides)]
  # rttt= [[row*sides+col+1 for col in range(sides)] for row in  range(sides)]
  # #its a reference tic tac toe to represent which number ca a player type in to mark that spot

  playerNo=1
  #lets start with player 1
  players = assign_player_codes()
  #players code has been defined now
  while positions_left:
    print(f"Turn for player no. {playerNo}")
    #playerNo is playing now
    display_ttt(ttt)
    display_rttt(ttt)
    r,c = get_unmarked_position(ttt,sides)
    # r and c are row and col to mark on the tic tac tow for current playerNo
    #Now mark the spot for playerNo:
    ttt[r][c]  = players[playerNo]
    #now we have to check if playerNo alreadyy won or not: if yes we will break out from the game and declare winner
    player_won= won_or_not(ttt,r,c,players[playerNo])
    if player_won:
      print(f"Player No. {playerNo} Won the Game")
      display_ttt(ttt)
      break
    else:
      print("Switching players")
      playerNo = switch_players(playerNo)
    positions_left-=1
  else:
    print("Game is a draw, No more positions left to fill")
    display_ttt(ttt)
```

```python
def askForReplay():
  choice= "wrong"
  while choice.lower() not in ['y','n']:
    choice = input("Wanna Playn again? press Y/y or if not press n/N")
    if choice.lower() not in ["y","n"]:
      print("Please try again form the given options")
  return choice.lower()=='y'
```

```python
replay = True
while replay:
  clear_output()
  play_ttt()
  replay = askForReplay()
```

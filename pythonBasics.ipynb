as of Python 3.7, you can absolutely rely on dictionaries being insertion-ordered. 
This means that when you iterate over a dictionary, the items will always appear in the exact order they were originally added
Updates vs. Re-insertions: If you update the value of an existing key, the order does not change. However, if you delete a key and then re-add it, that key moves to the end of the dictionary.
Reversible (Python 3.8+): Since the order is now stable, Python 3.8 added the ability to use the reversed() function on dictionaries and their views (keys, values, and items).

When to Still Use collections.OrderedDict
Even though standard dicts are now ordered, the collections.OrderedDict class still exists for three specific reasons: 
Order-Sensitive Equality: For standard dicts, {'a': 1, 'b': 2} == {'b': 2, 'a': 1} is True. For OrderedDict, this comparison is False because the order must match.
Explicit Reordering: OrderedDict has a .move_to_end() method that lets you move a key to either the beginning or the end efficiently.
Enhanced popitem(): In OrderedDict, you can use popitem(last=False) to remove and return the first item added; a standard dict's .popitem() always removes the last item.

While reversed() returns a "lazy" object (an iterator), sorted() returns a real, fully populated List immediately.

reversed() is Lazy: To reverse a list, Python just needs to know the last index and move backwards. It doesn't need to do any math or comparisons, so it gives you a "recipe" (iterator) to save memory.

sorted() is Eager: To sort a list, Python must compare every single element against the others (using the Timsort algorithm). It cannot "lazily" know which element is the smallest without looking at all of them. Therefore, it finishes the work and hands you a new list.

```python
def count_primes(num):
    if num < 2:
        return 0
    
    # 1. Create a boolean array "is_prime" and set all to True
    # We use num + 1 to include 'num' itself
    is_prime = [True] * (num + 1)
    is_prime[0] = is_prime[1] = False # 0 and 1 are not prime
    
    # 2. Only need to go up to sqrt(num)
    for p in range(2, int(num**0.5) + 1):
        if is_prime[p]:
            # 3. Mark all multiples of p as False
            # We start at p*p because smaller multiples are already marked
            for i in range(p * p, num + 1, p):
                is_prime[i] = False
                
    # 4. Sum the True values (True acts as 1 in Python)
    return sum(is_prime)
```

```python
#shallow copying

import copy

x = [1, 2, [3, 4]]
y = x.copy()  # Or y = x[:]

x[0] = "Changed" 
print(y[0])      # Prints 1 (y is safe!)

x[2][0] = "BOOM" 
print(y[2][0])   # Prints "BOOM" (y is NOT safe!)
```

```python
import copy

x = [1, 2, [3, 4]]
y = copy.deepcopy(x)

x[2][0] = "Safe"
print(y[2][0])   # Prints 3 (y is 100% independent!)
```

```python
import copy

# --- SCENARIO 1: The Basic Trap (Immutables) ---
a = 10
b = a
a = a + 5 

# --- SCENARIO 2: The Shallow vs Deep Battle ---
list_a = [1, 2, [3, 4]]
list_b = list_a[:]          # Shallow copy
list_c = copy.deepcopy(list_a) # Deep copy

list_a[0] = 99         # Changing an integer at the top level
list_a[2][0] = "BOOM"  # Changing a value inside the nested list

# --- SCENARIO 3: The Function/Global Confusion ---
counter = 100
my_list = [10, 20]

def mystery_func(c, l):
    c += 50            # c is an immutable int
    l.append(30)       # l is a mutable list
    l = [1, 1, 1]      # Reassigning l to a brand new list object

mystery_func(counter, my_list)

# --- THE TEST QUESTIONS ---
print(f"Q1: What is 'b'? Answer: {b}") 
# (Did b move with a, or stay at 10?)

print(f"Q2: list_b[0] is: {list_b[0]}") 
# (Did the shallow copy catch the change to the first number?)

print(f"Q3: list_b[2][0] is: {list_b[2][0]}") 
# (Did the shallow copy share the nested list 'BOOM'?)

print(f"Q4: list_c[2][0] is: {list_c[2][0]}") 
# (Did the deep copy stay safe as 3?)

print(f"Q5: counter is: {counter}") 
# (Did the function change the global integer?)

print(f"Q6: my_list is: {my_list}") 
# (Did the append work? Did the reassignment [1,1,1] work?)
```

```python
import copy

class Base:
    data = ["Base"]
    val = 1

    def __init__(self):
        self.hp = 100
        self.bag = [1, 2]

    def buff(self):
        # Pay close attention to what += does to a class attribute
        self.val += 10  
        self.hp += 50

class Sub(Base):
    data = ["Sub"]

    def __init__(self, init_parent):
        if init_parent:
            super().__init__()
        self.hp = 10
        self.nested = [[9]]

# --- Setup ---
obj_a = Sub(init_parent=True)
obj_b = Sub(init_parent=False)

# --- Actions ---

# 1. Class Attributes
Base.data.append("X")
obj_a.data.append("Y")

# 2. Methods and Instance Attributes
obj_a.val = 99
obj_b.buff()

# 3. Shallow Copying
obj_c = copy.copy(obj_a)
obj_c.bag.append(3)
obj_c.nested[0].append(8)
obj_c.val = 55

# 4. Function Scope
def tricky(o, lst):
    o.hp = 0
    lst.append("Z")
    lst = ["Reset"] 

tricky(obj_b, obj_b.data)

# --- The Test ---
print(Base.data)
print(Sub.data)
print(obj_a.data)
print(obj_a.val)
print(Base.val)
print(obj_b.hp)
print(hasattr(obj_b, "bag")) # (Does obj_b have the 'bag' attribute? True/False)
print(obj_b.val)
print(obj_a.bag)
print(obj_a.nested)
print(obj_c.val)
```

```python
import ctypes

class Book:
    def __init__(self, title):
        self.title = title
    def __del__(self):
        # Your custom goodbye message
        print(f"DEBUG: __del__ called for '{self.title}'")

# 1. Create the object and save its address
b1 = Book("Learning Python")
obj_id = id(b1)
print(b1)
print(f"Object created at memory address: {obj_id}")

# 2. Trigger your custom __del__
print("Running 'del b1' now...")
del b1

# 3. Prove it's gone
# Attempting to access freed memory is dangerous, but we can check 
# if Python's internal list of objects still contains this ID.
print(f"Post-deletion check for address {obj_id} completed.")
try:
    recovered_obj = ctypes.cast(obj_id, ctypes.py_object).value
    print(f"After del, I found this at the address: {recovered_obj}")
except Exception as e:
    print(f"Access Failed/Garbage found: {e}")
```

```python
def cool():
    def super_cool():
        return 'I am very cool!'
    print(f"id for super cool function inside cool fn is :{id(super_cool)}")
    return super_cool

# 1. First Run
some_func = cool()
id_first = id(some_func)
print(f"First object ID:  {id_first}")

# 2. Second Run (Re-assigning to the same variable)
some_func = cool()
id_second = id(some_func)
print(f"Second object ID: {id_second}")

# 3. Verification
if id_first != id_second:
    print("\nSUCCESS: The first object was replaced by a brand new one in a different memory location!")
```

```python
def cool(name):
    ageDefault = 25  
    def super_cool(age=ageDefault):
        # 'name' is used here, so it's kept in the "backpack"
        return f"Hello {name}! of age :{age}" 
    return super_cool

print(cool("Jassi Singh")(34))
print(cool("Jatin Singh")(26))
```

In this specific code, ageDefault is not saved in the "backpack" (closure). Instead, its value is hard-coded into the super_cool function object at the moment it is created.
Here is exactly what is happening in memory:
1. Default Arguments vs. Closures
The Default Argument (age=ageDefault): When Python defines super_cool, it looks up the value of ageDefault (which is 25) and stores that actual number in a special attribute of the function called __defaults__. It essentially "bakes" the number 25 into the function.
The Closure (name): Because name is used inside the body of the function, Python keeps the variable name alive in the __closure__ (the backpack).
2. Is ageDefault kept alive?
No. After cool("Jassi Singh") finishes its run:
The value 25 is already safe inside the super_cool function's default settings.
The variable name ageDefault itself is deleted from memory.
If you tried to look for ageDefault later, you wouldn't find it.
3. How to Prove it
You can inspect the function object to see where the data is stored:

```python
func = cool("Jassi Singh")

# Look at the "Backpack" (Closure) -> Only 'name' is here
print(f"Closure contents: {[c.cell_contents for c in func.__closure__]}") 
# Output: ['Jassi Singh']

# Look at the Default Arguments -> The number 25 is stored here
print(f"Default arguments: {func.__defaults__}") 
# Output: (25,)
```

```python
def cool(name):
    # This list is created every time cool() is called
    my_defaults = [] 
    # default argument trap
    def super_cool(items=my_defaults):
        items.append(name)
        return items
    
    return super_cool

# Case 1
f1 = cool("Jatin")
print(f1())  # ['Jatin']
print(f1())  # ['Jatin', 'Jatin'] -> IT PERSISTS for f1!

# Case 2
f2 = cool("Singh")
print(f2())  # ['Singh'] -> IT IS A FRESH LIST for f2!

print(f1())
print(f2())

print(f"f1 defaults: {f1.__defaults__}")
print(f"f2 defaults: {f2.__defaults__}")
print(f"backpack of f1: {[cell.cell_contents for cell in f1.__closure__]}")
print(f"backpack of f2: {[cell.cell_contents for cell in f2.__closure__]}")
print(f"f1 closure: {f1.__closure__}")
print(f"f1 closure's cell: {f1.__closure__[0]}")
print(f"f1 closure's cell's content : {f1.__closure__[0].cell_contents}")
```

```python
class MyClass:
    def __init__(self, val):
        self.val = val

obj = MyClass(10)
lst = [obj, obj, obj]

# Change the value through the first item in the list
lst[0].val = 99

# Everyone "changes"!
print(lst[1].val) # Output: 99
print(lst[2].val) # Output: 99
```

```python
print(lst[0] is lst[1]) # True (Exact same object in memory)
print(id(lst[0]) == id(lst[1])) # True (Their memory addresses are identical)
```

```python
# Three DIFFERENT objects with the same starting value
lst = [MyClass(10) for _ in range(3)]

lst[0].val = 99
print(lst[1].val) # Still 10!
```

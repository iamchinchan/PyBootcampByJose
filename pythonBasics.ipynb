```python
result = 10.65655
print("result is: {}".format(result))
print("result is: {:1.2f}".format(result))  # this also works, no need to name the var
print("result is: {r:6.2f}".format(r=result))
print("result is: {0:1.2f}".format(result)) # also works, as it takes 0 as index
print("result is: {0:1.2f}".format(result,55.52156))
print("result is: {1:1.2f}".format(result,55.52156))
```

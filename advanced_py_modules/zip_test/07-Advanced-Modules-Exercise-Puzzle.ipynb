```python
import re
pattern = r"\d{3}-\d{3}-\d{4}"
def findMatch():
    for folder , sub_folders,files  in os.walk('extracted_content/'):
        for file in files:
            with open(f"{folder}/{file}") as f:
                contents = f.read()
                # print(f"file is {folder}/{file}, and content is {contents}")
                match = re.search(pattern,contents)
                if match:
                    print(f"found match in file : {folder}/{file}")
                    return match
print(findMatch().group())
```

```python
path_to_walk  = "extracted_instructions\\extracted_content"
import os
import re
pattern = r"\d{3}-\d{3}-\d{4}"
found = False
for folder,sub_folders,files in os.walk(path_to_walk):
  # print(f"Currently in folder: {folder}")
  # print(f"It has sub folders: {sub_folders}")
  # print(f"it has files: {files}")
  for file in files:
    file_path = os.path.join(folder,file)
    with open(file_path,"r") as f:
      f.seek(0)
      file_content = f.read()
      result = re.search(pattern,file_content)
      if result:
        print(f"Found the phone number: {result.group()} in file: {folder+'\\'+file}")
        found = True
        break
      else:
        # print(f"Not found in file : {file}")
        pass
  if found:
    break
```

```python
from collections import namedtuple

# Define your "class"
Dog = namedtuple('Dog', ['age', 'breed', 'name'])

# Check the order
print(Dog._fields)
# Output: ('age', 'breed', 'name')
```

```python
from collections import namedtuple
Dog = namedtuple('Dog', ['age', 'breed', 'name'])
sammy = Dog(age=5, breed='Husky', name='Sammy')

# Create a new version of Sammy who is one year older
sammy = sammy._replace(age=6)

print(sammy.age) # Now it's 6!
#but i believe i shouldn not do this as it defied core purpose oif immmutable: but good to know i can: 
# but it creates a brand new object and old passed arguments might now point to iold adress so might cause inconsistency and confusion: althoughn thats whjat immutability means, it cant change 
```

## Complete list

For a complete list of the GUI widgets available to you, you can list the registered widget types. `Widget` is the base class.

```python
import ipywidgets as widgets
# Show all available widgets!
for item in widgets.Widget.widget_types.items():
    # print(item[0][2][:-5])
    print(item[0][2])
```

```python
from ipywidgets.widgets.widget import _widget_types

# Collect unique widget class names, sort them, and print
unique_widgets = sorted({cls.__name__ for cls in _widget_types.values()})

for name in unique_widgets:
    print(name)
```

## Numeric widgets

There are 10 widgets distributed with IPython that are designed to display numeric values. Widgets exist for displaying integers and floats, both bounded and unbounded.  The integer widgets share a similar naming scheme to their floating point counterparts. By replacing `Float` with `Int` in the widget name, you can find the Integer equivalent.

### IntSlider

```python
widgets.IntSlider(
    value=7,
    min=0,
    max=10,
    step=1,
    description='Test:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)
```

```python
import ipywidgets as widgets

def check_trigger(change):
    print(f"Python Kernel Woke Up! Value is now: {change['new']}")

# Test with False
slider = widgets.IntSlider(value=0, min=0, max=10, continuous_update=True)
slider.observe(check_trigger, names='value')
display(slider)
```

```python
import ipywidgets as widgets

# 1. Create a dedicated container to catch print statements
out = widgets.Output()

def check_trigger(change):
    # Clear the previous text output
    # out.clear_output()
    # Direct the print statement into our output container
    with out:
        print(f"Python Kernel Woke Up! Value is now: {change['new']}")

# Test with False
slider = widgets.IntSlider(value=0, min=0, max=10, continuous_update=True)
slider.observe(check_trigger, names='value')

# 2. Stack the slider and the output box vertically
widgets.VBox([slider, out])
```

```python

output = widgets.Text()
def change_output(trigger):
  output.value = str(trigger["new"])
slider = widgets.IntSlider(value=15,min=0,max=50,step=1,description="Int Slider: ",continuous_update = False, disabled = False, readout = True, readout_format="d")
slider.observe(change_output,names ="value")
widgets.VBox([slider,output])
```

## Selection widgets

There are several widgets that can be used to display single selection lists, and two that can be used to select multiple values. All inherit from the same base class. You can specify the **enumeration of selectable options by passing a list** (options are either (label, value) pairs, or simply values for which the labels are derived by calling `str`). You can **also specify the enumeration as a dictionary**, in which case the **keys will be used as the item displayed** in the list and the corresponding **value will be used** when an item is selected (in this case, since dictionaries are unordered: in latest verion they are ordered
, the displayed order of items in the widget is unspecified).

### Dropdown

```python
widgets.Dropdown(
    options=['1', '2', '3'],
    value='2',
    description='Number:',
    disabled=False,
)
```

The following is also valid:

```python
widgets.Dropdown(
    options={'One': 1, 'Two': 2, 'Three': 3},
    value=2,
    description='Number:',
)
```

```python
output = widgets.Output()
def update_output(trigger):
  output.clear_output()
  with output:
    print(f"user has selected : {trigger["new"]}")

options = widgets.Select(
    options=['Linux', 'Windows', 'OSX'],
    value='OSX',
    # rows=1,
    description='OS:',
    disabled=False
)
options.observe(update_output, names ="value")
widgets.VBox([options,output])
```

```python
output = widgets.Label()
# output = widgets.Output()
options = widgets.Select(
    options=['Linux', 'Windows', 'OSX'],
    value='OSX',
    # rows=1,
    description='OS:',
    disabled=False
)

my_link  =widgets.link((options,"value"),(output,"value"))
widgets.VBox([options,output])
```

### SelectionRangeSlider
The value, index, and label keys are 2-tuples of the min and max values selected. The options must be nonempty.

```python
import datetime
dates = [datetime.date(2015,i,1) for i in range(1,13)]
options = [(i.strftime('%b'), i) for i in dates]
widgets.SelectionRangeSlider(
    options=options,
    index=(5,10),
    description='Months (2015)',
    disabled=False
)
```

## String widgets
There are several widgets that can be used to display a string value. The `Text` and `Textarea` widgets accept input. The `HTML` and `HTMLMath` widgets display a string as HTML (`HTMLMath` also renders math). The `Label` widget can be used to construct a custom control label.

### Text

```python
widgets.Text(
    value='Hello World',
    placeholder='Type something',
    description='String:',
    disabled=False
)
```

### Textarea

```python
widgets.Textarea(
    value='Hello World',
    placeholder='Type something',
    description='String:',
    disabled=False
)
```

### Label
The `Label` widget is useful if you need to build a custom description next to a control using similar styling to the built-in control descriptions.

```python
widgets.HBox([widgets.Label(value="The $m$ in $E=mc^2$:"), widgets.FloatSlider()])
```

```python
import ipywidgets as widgets
from IPython.display import display, Markdown

# 1. Create a container to capture formatting updates
math_container = widgets.Output()

# 2. Direct the markdown compiler to render the formulas
with math_container:
    display(Markdown("The $m$ in $E=mc^2$:"))

# 3. Stack side-by-side inside your horizontal layout grid
widgets.HBox([math_container, widgets.FloatSlider()])
```

```python
import ipywidgets as widgets

# Use HTMLMath to trigger math parsing
widgets.HBox([
    widgets.HTMLMath(value="The $m$ in $E=mc^2$:"), 
    widgets.FloatSlider()
])
```

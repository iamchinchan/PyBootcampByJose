```python
for image in images:
  print(image["src"])
```

```python
kailasa_temple_image = None
Found=False
pattern  = "kailasa"
import re

for image in images:
  image_src = image["src"]
  #image_src is a str:
  Found = re.search(pattern.lower(),image_src.lower())
  if Found:
    #if True: Found anything
    kailasa_temple_image = image
    break 
else:
  print(f"Not Found Kailasa Temple Image")

if kailasa_temple_image:
  print(kailasa_temple_image)

    
```

```python
kailasa_temple_src = kailasa_temple_image["src"]
```

```python
kailasa_temple_src
```

```python
pattern_for_format = r"\.[a-zA-Z0-9]+$"
result = re.search(pattern_for_format,kailasa_temple_src)
# print(result)
kailasa_image_format = result.group()
```

```python
type(kailasa_image_format)
```

```python
kailasa_image_format
```

```python
binary_temple_image  = requests.get(f"https:{kailasa_temple_src}",headers= headers)
```

```python
binary_temple_image_content = binary_temple_image.content
```

```python
with open(f"kailasa_temple_image{kailasa_image_format}","wb") as f:
  f.write(binary_temple_image_content)
```
